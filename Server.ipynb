```python
#Server
print(">>>>>SERVER<<<<<" + "\n" + "\n")

import socket
# AF_INET == ipv4
# SOCK_STREAM == TCP
# 用socket.socket建立socket，socket.SOCK_STREA為TCP宣告
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
print("----------Socket Established----------" + "\n")

# ip = socket.gethostname() 為獲得主機名稱
ip = socket.gethostbyname(socket.gethostname())
# port number範圍從0~65535
port = 49496
server_address = (ip, port)
s.bind(server_address)
#設定可連接的socket數
s.listen(20)
print("----------Waiting For The Client's Connections----------" + "\n")


#接收client端的連接，取得client的IP位址並打招呼
while True:
    c, addr = s.accept()
    print("----------Here Is Client's Address : ", address, "----------" + "\n")
    serveruser_name = input("Please enter your name : " + "\n")
    c.send(bytes("Nice to meet you, my name is " + serveruser_name, "utf-8"))
    
#和client玩剪刀石頭布
    print(c.recv(1024).decode())
    
    server_input = input()
    if server_input == "0" :
        server_input1 = "scissors"

    elif server_input == "1" :
        server_input1 = "rock"

    elif server_input == "2" :
        server_input1 = "paper"

    else :
        server_input1 = "X"

    c.send(bytes(serveruser_name + " : (" + server_input + "," + server_input1 + ")" ,"utf-8"))
    print("----------Comparison----------")
    print(c.recv(1024).decode())
    print(serveruser_name + " : (" + server_input + "," + server_input1 + ")" + "\n")

    c.send(bytes(server_input, "utf-8"))
    client_input = c.recv(1024).decode()

    if (int(client_input) != 1) & (int(client_input) != 2) :
        if int(client_input) != 0 :
            print("Client input wrong number")
            print("----------Game End----------")

    if server_input == "0" :
        if client_input == "1" :
            print("Client Win!")
            print("----------Game End----------")

        elif client_input == "2" :
            print("Server Win!") 
            print("----------Game End----------")

        elif client_input == "0" :
            print("The game ended in a tie !")
            print("----------Game End----------")


    elif server_input == "1" :
        if client_input == "2" :
            print("Client Win!")
            print("----------Game End----------")

        elif client_input == "0" :
            print("Server Win!") 
            print("----------Game End----------")
        elif client_input == "1" :
            print("The game ended in a tie !")
            print("----------Game End----------")

    elif server_input == "2" :
        if client_input == "0" :
            print("Client Win!")
            print("----------Game End----------")

        elif client_input == "1" :
            print("Server Win!") 
            print("----------Game End----------")

        elif client_input == "2" :
            print("The game ended in a tie !")  
            print("----------Game End----------")

    else :
        print("Server input wrong number")
        print("----------Game End----------")
    c.close()
    print("----------Client Has Been Closed----------")




```
