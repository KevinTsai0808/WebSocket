```python
#Client
print(">>>>>CLIENT<<<<<" + "\n" + "\n")

import socket
import random 
# AF_INET == ipv4
# SOCK_STREAM == TCP
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

connecthost_ip = socket.gethostbyname(socket.gethostname())
connecthost_port = 49496
connecthost_address = (connecthost_ip, connecthost_port)
#使用connect連接想連接的host
s.connect(connecthost_address)
print("----------Connected Successfully----------" + "\n")

#和server打招呼
print(s.recv(1024).decode())
clientuser_name = input("Please enter your name : " + "\n") 
s.send(bytes("Hello, my name is " + clientuser_name + "\n" + "Let's play Rock Paper Scissors！" + "\n" + "Enter a number (0:scissors, 1:rock, 2:paper) : " , "utf-8"))

#和server玩剪刀石頭布
while True:
    client_input = input("Please enter a number (0:scissors, 1:rock, 2:paper) : " + "\n")
    if client_input == "0" :
        client_input1 = "scissors"

    elif client_input == "1" :
        client_input1 = "rock"

    elif client_input == "2" :
        client_input1 = "paper"

    else :
        client_input1 = "X"


    s.send(bytes(clientuser_name + " : (" + client_input + "," + client_input1 + ")","utf-8"))
    print("----------Comparison----------")
    print(s.recv(1024).decode())
    print(clientuser_name + " : (" + client_input + "," + client_input1 + ")" + "\n")
    s.send(bytes(client_input, "utf-8"))
    server_input = s.recv(1024).decode()

    if (int(client_input) != 1) & (int(client_input) != 2) :
            if int(client_input) != 0 :
                print("Client input wrong number")
                print("----------Game End----------")

    if server_input == "0" :
        if client_input == "1" :
                print("Client Win!")
                print("----------Game End----------")

        elif client_input == "2" :
                print("Server Win!") 
                print("----------Game End----------")

        elif client_input == "0" :
                print("The game ended in a tie !")
                print("----------Game End----------")


    elif server_input == "1" :
        if client_input == "2" :
                print("Client Win!")
                print("----------Game End----------")

        elif client_input == "0" :
                print("Server Win!") 
                print("----------Game End----------")
        elif client_input == "1" :
                print("The game ended in a tie !")
                print("----------Game End----------")

    elif server_input == "2" :
        if client_input == "0" :
                print("Client Win!")
                print("----------Game End----------")

        elif client_input == "1" :
                print("Server Win!") 
                print("----------Game End----------")

        elif client_input == "2" :
                print("The game ended in a tie !")  
                print("----------Game End----------")

    else :
            print("Server input wrong number")
            print("----------Game End----------")        
    break
print("----------Connection Ended----------")    




```
